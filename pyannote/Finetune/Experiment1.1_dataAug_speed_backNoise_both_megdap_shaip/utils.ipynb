# ploting the results of fine tune model on different dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Learning Rate": [0.001, 0.001, 0.001, 0.001, 0.0001, 0.0001, 0.0001, 0.0001],
    "Weight Decay": [0, 0, 0.005, 0.005, 0, 0, 0.005, 0.005],
    "Amsgrad": [False, True, False, True, False, True, False, True],
    "Shaip Test EER%": [0.0521, 0.0377, 0.0701, 0.0709, 0.0559, 0.0642, 0.0393, 0.0451],
    "Shaip Val EER%": [0.0514, 0.0383, 0.0635, 0.0632, 0.0741, 0.0624, 0.0346, 0.0603],
    "Megdap Test EER%": [0.0535, 0.0268, 0.0433, 0.06623, 0.0485, 0.0295, 0.0244, 0.0342],
    "Megdap Val EER%": [0.0683, 0.0284, 0.0353, 0.07174, 0.0523, 0.0388, 0.0317, 0.0428],
}

df = pd.DataFrame(data)
print(df)
df['Average EER%'] = df[['Shaip Test EER%', 'Shaip Val EER%', 'Megdap Test EER%', 'Megdap Val EER%']].mean(axis=1)

df 
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Average EER%'], 'o-', label='Average EER%')
plt.xlabel('Hyperparameter Setting')
plt.ylabel('Average EER%')
plt.title('Average EER% vs Hyperparameter Settings')
plt.xticks(ticks=range(len(df)), labels=[
    'LR=0.001, WD=0, Amsgrad=False',
    'LR=0.001, WD=0, Amsgrad=True',
    'LR=0.001, WD=0.005, Amsgrad=False',
    'LR=0.001, WD=0.005, Amsgrad=True',
    'LR=0.0001, WD=0, Amsgrad=False',
    'LR=0.0001, WD=0, Amsgrad=True',
    'LR=0.0001, WD=0.005, Amsgrad=False',
    'LR=0.0001, WD=0.005, Amsgrad=True'
], rotation=45, ha='right')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
